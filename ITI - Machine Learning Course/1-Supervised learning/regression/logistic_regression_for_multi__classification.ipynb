```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as plt
```

# 1.load dataset

```python
iris = load_iris()

df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target


df.head()
```

# 2. Data preprocessing

# 3.Data Exploration & Visualization

```python

plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'],c=df['target'] , marker='o')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title(" Iris Dataset")
plt.show()
```

# 4. select model

```python
from sklearn.model_selection import train_test_split
```

```python

X = df[iris.feature_names]
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# 5. model train

```python
from sklearn.linear_model import LogisticRegression

log_reg_multi = LogisticRegression(multi_class="multinomial", solver="lbfgs", max_iter=200)
log_reg_multi.fit(X_train, y_train)

```

# 6.evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Predictions
y_pred_multi = log_reg_multi.predict(X_test)

print("\nMulti-Class Classification Results (Iris):")
print('-'*50)
print("Accuracy:", accuracy_score(y_test, y_pred_multi)*100, "%")
print('-'*50)
print(confusion_matrix(y_test, y_pred_multi))
print('-'*50)
print(classification_report(y_test, y_pred_multi, target_names=iris.target_names))
print('-'*50)
```
